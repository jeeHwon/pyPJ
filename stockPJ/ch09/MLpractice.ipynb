```python
def AND(x1, x2):
    w1 = 0.5
    w2 = 0.5
    theta = 0.7
    if w1 * x1 + w2 * x2 > theta:
        return 1
    else :
        return 0
AND(0, 0)
AND(0, 1)
AND(1, 0)
AND(1, 1)
```

```python
def NAND(x1, x2):
    w1 = -0.5
    w2 = -0.5
    theta = -0.7
    if w1 * x1 + w2 * x2 > theta:
        return 1
    else:
        return 0
NAND(0, 0)
NAND(0, 1)
NAND(1, 0)
NAND(1, 1)
```

```python
def OR (x1, x2):
    w1 = 0.5
    w2 = 0.5
    theta = 0.2
    if w1 * x1 + w2 * x2 > theta:
        return 1
    else:
        return 0
OR(0, 0)
OR(0, 1)
OR(1, 0)
OR(1, 1)
```

```python
def XOR (x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))
XOR(0, 0)
XOR(0, 1)
XOR(1, 0)
XOR(1, 1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def stepfunc(x):
    return np.where(x<=0, 0, 1)

x = np.arange(-10, 10, 0.1)
y = stepfunc(x)

plt.plot(x, y)
plt.title('step function')
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-10, 10, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.title('sigmoid function')
plt.show()
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

x = np.arange(-10, 10, 0.1)
y = tanh(x)

plt.plot(x, y)
plt.title('tanh function')
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-10, 10, 0.1)
y = relu(x)

plt.plot(x, y)
plt.title('relu function')
plt.show()
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))

x = np.arange(-10, 10, 0.1)
y = softmax(x)

softmax([1, 1, 2])
```
